```python
# Combine all tags
import os
import json
import pandas as pd

root_dir = os.path.join("StackOverflow-Project", "questions")
sub_dirs = ["2013","2014", "2015", "2016", "2017"]

output_list = []

for sub_dir in sub_dirs:
    curr_dir = os.path.join(root_dir, sub_dir)
    all_files = os.listdir(curr_dir)
    for file in all_files:
        if file[-4:] == "json":
            f = open(os.path.join(curr_dir, file))
            data = json.load(f)
            for item in data["items"]:
                item_list = [item["question_id"], item["creation_date"], item["comment_count"], 
                             item["is_answered"], item["view_count"], item["answer_count"], 
                             item["score"], item["tags"]]
                output_list.append(item_list)
            f.close()
            
df = pd.DataFrame(output_list, columns = ["question_id", "creation_date", "comment_count", "is_answered", "view_count", "answer_count", "score", "tags"])
df.to_csv("all_tags.csv", index = False)
```

```python
import os
import json
import pandas as pd
import numpy as np

def get_item(item):
    item_list = []
    if "excerpt" in item:
        item_list =[item["excerpt_last_edit_date"], item["body_last_edit_date"], item["excerpt"], item["tag_name"]]
    else:
        item_list =[item["excerpt_last_edit_date"], item["body_last_edit_date"], np.nan, item["tag_name"]]
    return item_list


root_dir = os.path.join("StackOverflow-Project", "duplicatetag")
sub_dirs = ["1-1000","1001-2000", "2001-3000", "2001-3000", "3001-4000","4001-5000","5001-6000","6001-7000","7001-8000","8001-9000","9001-10000","10001-11000","11001-12000","12001-13000","13001-14000","14001-15000","15001-16000","16001-17000","17001-18000","18001-19000","19001-20000","20001-21000","21001-22000","22001-22108"]

output_list = []
count = 0
count_file = 0
count_no_item = 0
item_c = 0

for sub_dir in sub_dirs:
    print(sub_dir)
    curr_dir = os.path.join(root_dir, sub_dir)
    all_files = os.listdir(curr_dir)
    for file in all_files:
        if file[-4:] == "json":
            f = open(os.path.join(curr_dir, file))
            data = json.load(f)
            count_file += 1
            if "items" in data:
                count += 1
                for item in data["items"]:
                    item_list = get_item(item)
                    output_list.append(item_list)
                    item_c += 1
            else:
                count_no_item += 1
            f.close()
            
df = pd.DataFrame(output_list, columns = ["excerpt_last_edit_date", "body_last_edit_date", "excerpt", "tag_name"])
df.to_csv("tagwiki.csv", index = False)
print(count)
```
